```python
from PIL import Image

# Open the images
image1 = Image.open('/workspaces/7641HW3/kmeans_plots/Diabetes_KMeans_2Features.png')
image2 = Image.open('/workspaces/7641HW3/GMM_plots/Diabetes_GMM_2Features.png')

# Get the dimensions of the images
width1, height1 = image1.size
width2, height2 = image2.size

# Create a new image with a size that can fit both images side by side
combined_width = width1 + width2
combined_height = max(height1, height2)
combined_image = Image.new('RGB', (combined_width, combined_height))

# Paste the first image at (0, 0)
combined_image.paste(image1, (0, 0))

# Paste the second image next to the first one
combined_image.paste(image2, (width1, 0))

# Save the combined image
combined_image.save('/workspaces/7641HW3/3D_plots/Diabetes_2D_combined.png')

# Display the combined image (optional)
combined_image.show()
```
